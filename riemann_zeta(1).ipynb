# **Background info**

You have all probably at least heard of the famous function,

$$
\zeta(s) \ = \ \sum_{n=1}^{\infty} \frac{1}{n^{s}}.
$$

This is one of the most (maybe *the* most) significant functions in all of mathematics. Probably the main reason for this importance is the many relations to the prime numbers. Chiefly, Euler proved the alternative representation

$$
\zeta(s) \ = \ \prod_{p \,  \text{prime}} \frac{1}{1-p^{-s}}
$$

Of course, you've also probably heard about the famous "Riemann hypothesis". We'll get back to this later. 

### **Special cases**

Euler first proved the special case,

$$
\zeta(2) \ = \ \sum_{n=1}^{\infty} \frac{1}{n^{2}} \ = \  \frac{\pi^{2}}{6}.
$$

This is the "Basel problem". It was at the forefront of mathematics in the early 1700s. Euler worked it out in 1734 as one of his first notable achievements. 

Another useful special case is

$$
\zeta(4) \ = \ \frac{\pi^{4}}{90}.
$$

In general, for integer $k$, 

$$
\zeta(2k) \ = \ \frac{(-1)^{k+1} (2\pi)^{2k} B_{2k}}{2 (2k)!},
$$

where $B_{2k}$ are the even Bernoulli numbers. No simple formula is known for odd-integer arguments. But 

$$
\zeta(3) \ = \ 1.2020569031595942854\dots,
$$

which has the special name, *ApÃ©ry's constant.*

### **Convergence**

There's one major caveat to the original sum. It only converges for 

$$
\text{Re(s)} > 1.
$$

There is another problem. Even for $\text{Re}(s) > 1$ the convergence can be *very slow*.


### **Alternating series acceleration**

We can use a clever trick to improve things a bit. Notice for any *convergent* sum 

$$
\sum_{n=1}^{\infty} \ = \ \sum_{n \, \text{odd}}  \ + \ \sum_{n \, \text{even}}.
$$

We need to use the qualifier *convergent*. It's not true in general that we can rearrange the terms in an infinite sum. But let's ignore the fine print for now as long as things are working for us. 

Going a step further, for any convergent sum,

$$
\sum_{n=1}^{\infty}a_{n} \ = \ \sum_{n=1}^{\infty} (-1)^{n+1} a_{n}  \ + \ 2\sum_{n=1}^{\infty} a_{2n}.
$$

Here's the crux: 

$$
a_{n} \ = \ \frac{1}{n^{s}} \qquad \implies \qquad a_{2n} \ = \ \frac{a_{n}}{2^{s}}.
$$

This means that for the Riemann zeta function 

$$
\zeta(s) \ = \ \frac{1}{1-2^{1-s}} \sum_{n=1}^{\infty} \frac{(-1)^{n+1}}{n^{s}}.
$$

We've written the original series in terms of an *alternating series*! This will give the same answer where the original version converged. But it also happens to converge for 

$$
\text{Re}(s) > 0, \qquad \text{provided} \qquad 2^{1-s} \ \ne \ 1.
$$


### **The Riemann hypothesis**

Aside from popular appeal, the Riemann hypothesis (RH) is notable because it's equivalent to a vast number of major pure mathematical results. Many theorems are "proven", assuming the hypothesis is true. If it's false, a lot of work will fall alongside. What is the RH? 

First, when dealing with complex-valued inputs, it's common to define 

$$
s = \sigma + i t,
$$

where $\sigma = \text{Re}(s)$, and $t = \text{Im}(s)$. The alternating series converges for $\sigma > 0$. The RH is the following 

$$
\zeta(s) \ = \ 0 \qquad \iff \qquad s = \frac{1}{2} + i \, t,
$$

We can also extend the zeta function to negative inputs. In that case, there exists another simple family of zeros. But *at this point*, the only zeros are expected to have a real part of *exactly* 1/2. It is known that all the non-trivial zeros lie in what's called the "critical strip" where $0 < \text{Re}(s) < 1$. This fact is also equivalent to the "Prime Number Theorem", which is obviously very impressive. So it's a bit of a giveaway that the RH is closely linked to lots of important math.

There's still a big problem with the function as we've currently defined it. While technically the series converges for (e.g.) $\sigma = 1/2$, it does so ***ridiculously, tediously, slowly***. For example, 

$$
\sum_{n = 1}^{\infty} \frac{(-1)^{n+1}}{\sqrt{n}} \ = \ 0.60489864342163037025...
$$

But getting only **3 digits** (0.604) takes over **one million** terms in the sum! 

If we want to find any zeros of the zeta function, we are going to need some better tricks. 

<br>

## **Euler summation**

Euler was a smart guy. He figured out a way to seriously *accelerate* alternating sums. It works by using *finite differences*. 


Start with a sequence: 

$$
a \ = \ (a_{0}, \, a_{1}, \, a_{2},\, \ldots)
$$

Compute the first *forward finite difference* 

$$
\Delta a \ = \ (a_{1}-a_{0}, \, a_{2}-a_{1}, \, a_{3}-a_{2},\, \ldots).
$$

We can repeat the operation

$$
\Delta^{2} a \ = \ (\Delta a_{1}-\Delta a_{0}, \, \Delta a_{2}-\Delta a_{1}, \, \Delta a_{3}-\Delta a_{2},\, \ldots).
$$

In terms of the original sequence, 

$$
\Delta^{2} a \ = \ ( a_{2} - 2 a_{1} + a_{0}, \, a_{3} - 2 a_{2} + a_{1}, \, a_{4} - 2 a_{3} + a_{2},\, \ldots).
$$

And

$$
\Delta^{3} a \ = \ ( a_{3} - 3 a_{2} + 3 a_{1} - a_{0}, \, a_{4} - 3 a_{3} + 3 a_{2} - a_{1}, \, a_{5} - 3 a_{4} + 3 a_{3} - a_{2},\, \ldots).
$$

We can keep going. In each case, we can take the 0-th term in the $n$th finite-difference sequence,

$$
b_{n} \ = \ (\Delta^{n} a)_{\,0}.
$$

There's an explicit formula for the result

$$
b_{n} \ = \ \sum_{k=0}^{n} \binom{n}{k} (-1)^{k} a_{n-k}.
$$

But even though there's a "closed-form" expression, it's best to compute $b_{n}$ directly, differencing the sequence multiple times. 


The main result is that for any convergent alternating series,

$$
\sum_{n=0}^{\infty} (-1)^{n} a_{n} \ = \ \sum_{n=0}^{\infty} \frac{(-1)^{n}}{2^{n+1}} (\Delta^{n} a)_{\,0}.
$$

Proving it isn't very hard. It mainly requires the good-old "reversing-the-sums trick":

$$
\sum_{n=0}^{\infty}\sum_{k=0}^{n} \ = \ \sum_{n=0}^{\infty}\sum_{k=0}^{\infty} \text{If}\,( k \le n)  \ = \ \sum_{k=0}^{\infty} \sum_{n=k}^{\infty}.
$$

After this, the only thing that's need is 

$$
\sum_{k=n}^{\infty} \binom{n}{k} \frac{1}{2^{n+1}} \ = \ 1,
$$

which you can prove with some effort by differentiating a geometric series $k+1$ times.

Overall, Euler summation is a *weird* result. 

* The condition for convergence is only that $\lim_{n\to \infty} a_{n} = 0$.


* The finite differences will usually decrease even faster. 


* The $2^{n+1}$ in the denominator means that the series actually converges *geometrically*, even if the original series is rubbish. 


* In the case of $s=1/2$, we can get **5 correct digits** using only **ten** terms in the Euler sum; it would take roughly *ten billion* terms in the original alternating version of the series.


* Finally, the Euler sum can converge in cases where it's absurd for the original series to converge. For example when $s < 0$.  

Putting everything together,

$$
\zeta(s) \ = \ \frac{1}{1-2^{1-s}}\sum_{n = 0}^{\infty} \frac{1}{2^{n+1}} \sum_{k=0}^{n}\binom{n}{k} \frac{(-1)^{k}}{(k+1)^{s}}
$$

It turns out this is a **globally converging series**. It works for any $s$ in the complex plane, provided $2^{1-s} \ne 1$. 

**But be careful!** This series is appealing in closed-form. But it's best to think of it as an Euler summation accelerated alternating series. The binomial coefficients become very large, e.g.,

$$
\binom{50}{25} \ = \ 126410606437752.
$$

So ***do not*** try to compute the inner sum directly. Compute it by *differencing* the original series. 


### **Extra special cases**

We can now try the series with $s < 0$. For example, $s=-1$ and 

$$
a_{n} = n \qquad (\Delta a)_{n} \ = \ 1, \qquad (\Delta^{2} a)_{n} \ = \ 0 \quad \text{for} \quad n > 1.
$$

After a finite number of terms, the differencing vanishes. We can put everything together and find the famous results

$$
\zeta(0) \ = \ -\frac{1}{2}, \qquad \zeta(-1) \ = \ -\frac{1}{12}.
$$

More generally, for integer $k$,

$$
\zeta(-k) \ = \ \frac{(-1)^{k}B_{k+1}}{k+1},
$$

where we again find the Bernoulli numbers.

Negative even integers are also roots for the zeta function on the full complex plane,

$$
\zeta(-2k) \ = \ 0.
$$

The full statement of the Riemann hypothesis excludes these "trivial zeros".

Finally, the zeta function has a proper "pole" at $s=1$. But even then, it's possible to get something useful. The "principle value" at $s=1$ is 

$$
\lim_{\varepsilon \to 0} \frac{\zeta(1+\varepsilon)+\zeta(1-\varepsilon)}{2} \ = \ \gamma \ = \ 0.57721566490153286061\ldots,
$$

where $\gamma$ is the Euler-Mascheroni constant. Getting the principal value correct would be one of the most difficult tests for any numerical implementation. It's quite hard to get more than 7-9 decimal digits. 



### **Reflection**

There is one more caveat to everything so far. The formula for the Euler summation is technically valid for the whole complex plane. But it starts behaving badly for large negative values of $\text{Re}(s)$.

There is one more magical formula to the rescue, 

$$
\zeta(s) \ = \ 2^{s} \pi^{1-s} \sin\left( \frac{\pi s}{2}\right) \,\Gamma(1-s) \,\zeta(1-s),
$$


The reflection formula works everywhere in the complex plane. You can use it to compute $\zeta(s)$ for (e.g.) $\text{Re}(s) < -5$. 

The famous $\Gamma$ function is 

$$
\Gamma(s) \ = \ \int_{0}^{\infty} x^{s-1} e^{-x} \,\text{d}x 
$$

Also, there is another special function called *digamma* that is defined such that 

$$
\psi(s) \ = \ \frac{\Gamma'(s)}{\Gamma(s)} \ = \ \frac{d}{ds} \log \Gamma(s).
$$

The non-trivial roots of the Riemann zeta function are symmetric across the real axis. That means that 

$$
s \ = \ \frac{1}{2} \pm i\, t
$$

are both roots if one of them is. 

We can use Newton's method to find roots of the zeta function given a reasonable initial guess 

$$
s_{0} \ = \ \frac{1}{2} \pm i\, t_{0}, \qquad \text{where} \qquad t_{0} > 0.
$$


That is 

$$
s_{n+1} \ = \ s_{n} - \frac{\zeta(s_{n})}{\zeta'(s_{n})}.
$$

The roots are distributed more-or-less randomly ascending along the $s=1/2$ axis. Exactly how random is the subject of much speculation. The 1st four roots have

$$
t \quad \approx \quad 14.13472514173469, \quad 21.02203963877155, \quad 25.01085758014568, \quad 30.42487612585951,
$$

where each number shows 16 correct digits.


After that, the roots tend to be spaced a few units apart as far as you care to look. There are 29 roots with $t < 100$. The 29th root has 

$$
